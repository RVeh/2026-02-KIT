## WILSON-KI mit Konfidenzellipse

### Ziel dieses Notebooks

Dieses Notebook dient dazu, den Zusammenhang zwischen
- Prognoseintervallen für Stichprobenergebnisse
- und Konfidenzintervallen für einen unbekannten Parameter

**geometrisch und gedanklich** zu klären, wobei die folgende Definition eines Konfidenzintervalls umgesetzt wird:
> **Definition:**  
> Ein Konfidenzintervall zu einer beobachteten relativen Häufigkeit besteht aus all denjenigen
> Wahrscheinlichkeiten im Bin($n,p$)-Modell, in deren Prognoseintervall die beobachtete relative Häufigkeit $h$ liegt.


Der Schnitt des oberen Astes der Ellipse mit $h$ liefert die linke Grenze $p_u$ des Konfidenzintervalls:
$$ h = p_u + z\sqrt{\frac{p_u(1-p_u)}{n}}. $$

Der Schnitt des unteren Astes der Ellipse mit $h$ liefert die rechte Grenze $p_o$ des Konfidenzintervalls:
$$ h = p_o - z\sqrt{\frac{p_o(1-p_o)}{n}}. $$

### Was kann man hier verändern?
- `n`: Stichprobengröße  
- `gamma`: Konfidenzniveau  
- `h`: Stichprobenergebnis
- `k`: Anzahl der sichtbaren Prognoseintervalle im KI

### Wichtige Hinweise

- Die Wiederholung findet **nur im Modell**, nicht in der Realität statt.
- Die Darstellung ist keine Simulation von Daten,
  sondern eine **Visualisierung eines statistischen Verfahrens**.
