## Simulation von Konfidenzintervallen (Wilson)

In diesem Notebook wird das Verhalten von Konfidenzintervallen
für einen unbekannten Anteilsparameter untersucht.

Wir arbeiten **im $B(n;p)$-Modell**:
- $n$ ist die Stichprobengröße,
- $p$ ist der (unbekannte) Modellparameter,
- $h = x/n$ ist die beobachtete relative Häufigkeit einer Stichprobe.

Zu jeder simulierten Stichprobe wird ein **Konfidenzintervall** berechnet.
Dabei verwenden wir das **Wilson-Verfahren**.

Wichtig:
- In der Realität haben wir **nur eine Stichprobe** und damit **nur ein Intervall**.
- Die Wiederholung vieler Stichproben findet **nur im Modell** statt.
- Die Simulation dient dazu zu verstehen, **was ein Konfidenzverfahren leistet**.

Die Grafik zeigt:
- jedes Konfidenzintervall als waagrechte Strecke,
- den wahren Parameterwert $p$ als senkrechte Linie,
- Intervalle, die $p$ überdecken (blau),
- Intervalle, die $p$ nicht überdecken (rot).

Die angegebene Trefferquote beschreibt,
wie oft das Verfahren bei Wiederholung im Modell den Parameter überdeckt.


```python
from core.wilson_core import simulate_wilson_intervals
from plot.wilson_plot import plot_intervals


# ============================================================
# Parameter (hier dürfen SuS/LuL experimentieren)
# ============================================================
n = 80
p_true = 0.60
gamma = 0.95
m = 100

seed = 7   # Empfehlung: feste Zahl für reproduzierbare Diskussion
# seed = None  # bewusst: jedes Mal neue Simulation


intervals, cover, rate = simulate_wilson_intervals(
    n=n,
    p_true=p_true,
    gamma=gamma,
    m=m,
    seed=seed
)

title = (
    f"{m} Intervalle (Wilson), n={n}, γ={gamma:.2f}\n"
    f"Trefferquote ≈ {rate:.2f}  |  Seed {seed}"
)

plot_intervals(
    intervals=intervals,
    cover=cover,
    p_true=p_true,
    title=title,
    show=True,
    outpath=None
)

print("Trefferquote:", round(rate, 3))
```
